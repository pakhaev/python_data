```python
import re
# 'p' point 1
def search_holder(documents, number):
    name = ''
    for i in documents:
        if number == i['number']:
            name = i['name']
    if name == '':
        name = -1
    return name
        
# 's' point 2
def search_direct(directories, number):
    direct = 0
    for i in directories:
        for j in directories[i]:
            if number == j:
                direct = int(i)
    if direct == 0:
        direct = -1
    return direct

# 'l' point 3
def print_documents(documents, directories):
    print("Текущий список документов:")
    for i in documents:
        print ("№: %s, mun: %s, владелец: %s, полка хранения: %d" %\
               (i['number'], i['type'], search_holder(documents, i['number']), \
                search_direct(directories, i['number'])))

#START other functions
# print current direct.
def print_directories(directories):
    print("Текущий перечень полок:", end=" ")
    last = list(directories)[-1]
    for i in directories:
        if i == last:
            print(i, end=".\n")
        else:
            print(i, end=", ")

# for input integer
def input_num(str):
    try:
        str += '\n'
        number = int(input(str))
        return number
    except:
        print("Проверьте правильность введенных цифр!")
        return -1

# found directories
def ad_found_dir(directories, number):
    count = ''
    for i in directories:
        if str(number) == i:
            count = i
            break
    if count == '':
        return 0
#END other functions

# 'as' point 4
def add_directories(directories):
    number = input_num("Введите номер полки:")
    if number == -1:
        return
    if str(number) in directories:
        print("Такая полка уже существует.", end=" ")
    elif number != '':
        directories[number] = []
        print("Полка добавлена.", end=" ")

# 'ds' point 5:
def delete_directories(directories):
    number = input_num("Введите номер полки:")
    if number == -1:
        return
    count = ''
    len_dir = -1
    if number == -1:
        return
    for i in directories:
        if str(number) == i:
            count = i
            len_dir = len(directories[i])
    if len_dir == 0:
        directories.pop(str(count))
        print("Полка удалена.", end=" ")
    elif len_dir > 0:
        print("На полке есть документа, удалите их перед удалением полки.", end=" ")
    elif count == '':
        print("Такой полки не существует.", end=" ")
    print_directories(directories)

# 'ad' point 1 / task 2
def add_documents(documents, directories):
    doc_number = input("Введите номер документа:\n")
    doc_type = input("Введите тип документа:\n")
    doc_holder = input("Введите владельца документа:\n")
    doc_dir = input_num("Введите полку для хранения:")
    if doc_dir == -1:
        return
    if ad_found_dir(directories, doc_dir) == 0:
        print("Такой полки не существует. Добавьте полку командой as.")
    else:
        directories[str(doc_dir)].append(doc_number)
        documents.append({'type': doc_type, 'number': doc_number, 'name': doc_holder})
    print_documents(documents, directories)

# delete document in directories
def d_doc_in_dir(documents, directories, number):
    dir_num = search_direct(directories, number)
    for i in range(len(directories[str(dir_num)])):
        if number == directories[str(dir_num)][i]:
            del directories[str(dir_num)][i]

# 'd' point 2 / task 2
def delete_document(documents, directories):
    number = input("Введите номер документа:\n")
    position = -1
    flag = 0
    for i in documents:
        position += 1
        if number == i['number']:
            flag = 1
            break
    if flag == 0:
        print("Документ не найден в базе.", end="")
    else:
        d_doc_in_dir(documents, directories, number)
        del documents[position]
    print_documents(documents, directories)
    
# 'm' point 3 / task 2
def move_document(documents, directories):
    number = input("Введите номер документа:\n")
    doc_dir = input_num("Введите номер полки:")
    flag = 0
    for i in documents:
        if number == i['number']:
            flag = 1
    if flag == 0:
        print("Документ не найден в базе.")
        print_documents(documents, directories)
    elif ad_found_dir(directories, doc_dir) == 0:
        print("Такой полки не существует.", end="")
        print_directories(directories)
    else:
        d_doc_in_dir(documents, directories, number)
        directories[str(doc_dir)].append(number)
        print ("Документ перемещен.")
        print_documents(documents, directories)

# currect input command
def correct_command(documents, directories, command):
    if command == 'p':
        number = input("Введите номер документа:\n")
        result = search_holder(documents, number)
        if result == -1:
            print("Документ не найден в базе")
        else:
            print("Владелец документа:", result)
    elif command == 's':
        number = input("Введите номер документа:\n")
        result = search_direct(directories, number)
        if result == -1:
            print("Документ не найден в базе")
        else:
            print("Документ хранится на полке:", result)
    elif command == 'l':
        print_documents(documents, directories)
    elif command == 'as':
        add_directories(directories)
        print_directories(directories)
    elif command == 'ds':
        delete_directories(directories)
    elif command == 'ad':
        add_documents(documents, directories)
    elif command == 'd':
        delete_document(documents, directories)
    elif command == 'm':
        move_document(documents, directories)
    else:
        print("Incorrect command!")

documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}

while (1):
    command = input("Введите команду:\n")
    if command == 'q':
        break
    correct_command(documents, directories, command)
```
