```python
Смоделировать игру против лотерейного автомата типа “777”. 
Игрок платит 1 руб., после чего выпадает случайное целое число, 
равномерно распределенное от 0 до 999. При некоторых значениях 
числа игрок получает выигрыш.

777: 200 руб.
999: 100 руб.
555: 50 руб.
333: 15 руб.
111: 10 руб.
*77: 5 руб.
**7: 3 руб.
*00: 2 руб.
**0: 1 руб.
    
Вопросы:

    Выгодна ли игра игроку?
    Сколько в среднем приобретает или теряет игрок за одну игру?
```

```python
from random import randint
from math import factorial

def win_count(win_numbers, max_number):
    win = 0
    for i in range(0, max_number, 1):
        if i in win_numbers or \
            i % 100 in win_numbers or \
            i % 10 in win_numbers or \
            str(i)[-2:] in win_numbers or \
            str(i)[-1:] in win_numbers:
            win += 1
    return win

max_number = 999
win_numbers = {777: 200, 999: 100, 555: 50, 333: 15, 111: 10, 77: 5, 7: 3, '00' : 2, '0': 1}
strl = ""
count = 10 #количество игр
money = 100 #деньги на начало игры

countWinNumbers = win_count(win_numbers, max_number)

print("Выигрышные комбинации:", countWinNumbers)
print()
print("Вероятность выигрыша:", countWinNumbers / max_number)
print("Вероятность проигрыша:", (max_number - countWinNumbers) / max_number)

mayWin = 0
for i in win_numbers:
    mayWin += win_numbers[i]
print("В среднем можно выиграть: ", round(mayWin / len(win_numbers), 2))

print()
win = 0
winCount = 0
failureCount = 0
totalWin = 0
print("Деньги на начало игры: %dруб." % (money))
for _ in range(count):
    money -= 1
    if(money < 0): #Закоментируй это, если денег столько же, сколько count и не боишься уйти в минус
        print("У вас на счету закончились деньги")
        money = 0
        break
    strl = str(randint(0, 9)) + str(randint(0, 9)) + str(randint(0, 9))
    if len(strl) != 3:
        print("Автомат сломался, пересядьте за другой")
        money += 1
        break
    rand_number = int(strl)
    # print("\nВыпавшее число:", strl)
    if rand_number in win_numbers:
        win = win_numbers[rand_number]
    elif rand_number % 100 in win_numbers:
        win = win_numbers[rand_number % 100]
    elif rand_number % 10 in win_numbers:
        win = win_numbers[rand_number % 10]
    elif strl[-2:] in win_numbers:
        win = win_numbers[strl[-2:]]
    elif strl[-1:] in win_numbers:
        win = win_numbers[strl[-1:]]
    else:
        # print("Вы не выиграли -1 руб\n")
        failureCount += 1
        continue
    winCount += 1
    totalWin += win
    money += win
    # print("Ваш выигрыш составил: " + str(win) + " руб\n")

print("За %d игр ваш счет может стать: %dруб." % (count, money))

print("Число выигрышей: %d на сумму: %dруб. | Число проигрышей: %d" % (winCount, totalWin, failureCount))
```
