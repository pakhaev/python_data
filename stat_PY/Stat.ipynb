```python
import pandas as pd

# Задание первое
data = pd.read_csv("horse_data.csv", sep=";", na_values=['?'])
horses = pd.DataFrame()
horses['surgery?'] = data.iloc[: , 0]
horses['Age'] = pd.to_numeric(data.iloc[: , 1])
horses['rectal_temperature'] = data.iloc[: , 3]
horses['pulse'] = data.iloc[: , 4]
horses['respiratory_rate'] = data.iloc[: , 5]
horses['temperature_of_extremities'] = data.iloc[: , 6]
horses['pain'] = data.iloc[: , 10]
horses['outcome'] = data.iloc[: , 22]
horses.to_csv('horses.csv', index=False)
horses.head(10)
```

```python
# Задание второе
print("Средний возраст:", horses.Age.mean())
print("Средний пульс:", horses['pulse'].mean())
print("Средняя ректальная температура:", horses['rectal_temperature'].mean())
print("Средняя частота дыхания:", horses['respiratory_rate'].mean())
print("Медиана частоты дыхания:", horses['respiratory_rate'].median())
print("Выброс частоты дыхания:", horses['respiratory_rate'].mean() - horses['respiratory_rate'].median())
```

```python
print(horses.isna().sum())


horses['surgery?'].fillna(horses['surgery?'].median(), inplace=True)

# В зависимости от возраста расчитываю среднюю температуру
horses['rectal_temperature'].fillna(horses.groupby('Age')\
                                    ['rectal_temperature'].transform('mean'), inplace=True)
# В зависимости от возраста расчитываю среднюю частоту дыхания
horses['respiratory_rate'].fillna(horses.groupby(['Age'])\
                                    ['respiratory_rate'].transform('mean'), inplace=True)
# Исходя из возраста и температуры рассчет медианы пульса
horses['pulse'].fillna(horses.groupby(['Age', 'rectal_temperature'])\
                                    ['pulse'].transform('median'), inplace=True)
# По частоте дыхания
horses['pain'].fillna(horses.groupby(['respiratory_rate'])\
                                    ['pain'].transform('median'), inplace=True)
# По температуре тела
horses['temperature_of_extremities'].fillna(horses.groupby(['rectal_temperature'])\
                                    ['temperature_of_extremities'].transform('median'), inplace=True)

# Заполняю оставшиеся пропуски медианами
horses['temperature_of_extremities'].fillna(horses['temperature_of_extremities'].median(), inplace=True)
horses['pain'].fillna(horses['pain'].median(), inplace=True)
horses['outcome'].fillna(horses['outcome'].median(), inplace=True)

horses['rectal_temperature'] = horses['rectal_temperature'].round(1)
horses['respiratory_rate'] = horses['respiratory_rate'].round(1)
horses['pulse'] = horses['pulse'].round(1)
print(horses.isna().sum())

horses.to_csv('output.csv', index=False)
```
